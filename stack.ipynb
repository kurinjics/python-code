```python
def push(stk):
    ele = int(input("Enter the integer to be pushed:"))
    stk.append(ele)
def pop(stk):
    if stk == []:
        print("Stack Underflow")
    else:
        print("Popped element from the stack is:",stk.pop())
def display(stk):
    if stk ==[]:
        print("Stack Underflow")
    else:
        top = len(stk)-1
        print(stk[top],"Top")
        for i in range(top-1,-1,-1):
            print(stk[i])
stk = []
top = None
print("1. Push")
print("2. Pop")
print("3. Display")
print("4. Exit")
while True:
    ch = int(input("Enter the choice:"))
    if ch == 1:
        push(stk)
    elif ch == 2:
        pop(stk)
    elif ch == 3:
        display(stk)
    elif ch == 4:
        break
    else:
        print("Enter the correct option")

        
```

```python
import csv
def write():
    f = open("details.csv","w",newline= '\r\n')
    wo = csv.writer(f)
    wo.writerow(["UserId","Password"])

    while True:
        u_id = input("Enter User - Id:")
        pswd = input("Enter Password:")
        data=[u_id,pswd]
        wo.writerow(data)
        ch = input("Do you want to enter more record(Y/N):")
        if ch in "nN":
            break
    f.close()
def read():
    f = open("details.csv","r",newline="\r\n")
    ro = csv.reader(f)
    for i in ro:
        print(i)
    f.close()
def search():
    f = open("details.csv","r",newline="\r\n")
    u = input("Enter user -id to search:")
    ro = csv.reader(f)
    next(ro)
    for i in ro:
        if i[0]==u:
            print(i[1])
            break
    else:
            print("Not Found")
write()
read()
search()
```
